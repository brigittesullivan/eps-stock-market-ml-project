```python
result_eps = pd.DataFrame()
def get_basic_eps(row):  
    result = pd.DataFrame()
    result = pd.DataFrame(columns = ['Earnings Date', 'ticker', "EPS Estimate", "Reported EPS", "Surprise(%)"])
    print(row)
    return result
    try:
        ticker_data = yf.Ticker(row)
        eps_data  = ticker_data.get_earnings_dates(limit=12)
        # income_stmt = ticker_data.income_stmt.T["Basic EPS"]
        result = pd.DataFrame(eps_data).reset_index()
        result['ticker'] = row
        result = result[['Earnings Date', 'ticker', "EPS Estimate", "Reported EPS", "Surprise(%)"]].copy()
        result.dropna(inplace = True)
        pd.concat(result_eps, result)
        return result
        
    except Exception as e:
        print(f'Error:{e}')
        return np.nan
```

```python
def check_ticker(row):
    """for each ticker, get earnings data if no earnings data found for the stock, return the error and the symbol"""
    errors_list = [] # to keep track of what happened to each stock
    result = pd.DataFrame()
    try:
        ticker_data = yf.Ticker(row)
        eps_data  = ticker_data.get_earnings_dates(limit=12).reset_index()
        eps_data['Symbol'] = row # add a column with the symbol into the DF (to concat later)
        # summary_list.append({'Symbol': row, 'Errors': 'Earnings Data found'})
        result = eps_data
    except Exception as e:
        result = {'Symbol': row, 'Errors': f"{e}"}
    res_dict = [result, errors_list]
    return result
```

Warning: the below cell can take a long time to run.

```python
import timeit

start_time = timeit.default_timer()

# get earnings data for each ticker: 
tuple = unique_tickers['ticker'].apply(check_ticker)

## Time:
elapsed_seconds = timeit.default_timer() - start_time
elapsed_minutes = elapsed_seconds / 60
print(f"Time taken: {elapsed_seconds} seconds")
print(f"Time taken: {elapsed_minutes} minutes")
```
